# Comparing to RCP 8.5

RCP 8.5 is the CMIP5 high emissions scenario, defined by the radiative forcing.  Since it represents a kind of "business as usual" scenario outside of PAGE, it provides a reasonable comparison case for the radiative forcings produced by PAGE.

```julia
# Load the RCP 8.5 radiative forcings (by gas)
using DataFrames

radforcing = readtable("../../calibration/rcp85-radforcing.csv")
```

```julia
# Run the model
include("../../src/main_model.jl")
```

Here we produce a comparison dataframe, between `estimated` radiative forcings for CO2, N2O, and CH4, and those under RCP 8.5 (labeled by `rcp85`).

```julia
# Pull out the particular years
rows = convert(Vector{Bool}, map(year -> in(year, m.indices_values[:time]), radforcing[:, 1]))

# CO2
dfco2 = DataFrame(co2_estimated=m[:co2forcing, :f_CO2forcing], co2_rcp85=radforcing[rows, :CO2_RF])

# CH4
dfch4 = DataFrame(ch4_estimated=m[:ch4forcing, :f_CH4forcing], ch4_rcp85=radforcing[rows, :CH4_RF])

# N2O
dfn2o = DataFrame(n2o_estimated=m[:n2oforcing, :f_N2Oforcing], n2o_rcp85=radforcing[rows, :N2O_RF])

df = hcat(DataFrame(year=m.indices_values[:time]), dfco2, dfch4, dfn2o)
```

```julia
using RCall
R"library(ggplot2)"
R"library(reshape2)"
R"df = melt($df id.vars='year')"
R"df$gas = substr(df$variable, 1, 3)"
R"df$source = substring(df$variable, 5)"
R"ggplot(df, aes(x=year, y=value, colour=source)) + facet_grid(gas ~ .) + geom_line()"
```
